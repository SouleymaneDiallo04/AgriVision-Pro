# **AgriVision Pro : Réinventer l'Agriculture grâce à l'IA**

R.jpeg

# **Présentation**

AgriVision Pro est une application innovante qui utilise l'intelligence artificielle pour améliorer la gestion des cultures, prédire les rendements et détecter les maladies des plantes. Conçue pour les agriculteurs et les chercheurs, elle permet une surveillance en temps réel des cultures, avec des recommandations basées sur des données précises pour optimiser les récoltes et garantir une agriculture plus durable.

# **Description du Projet**

AgriVision Pro est conçu pour proposer des solutions efficaces à trois des plus grands défis de l'agriculture : Recommandation de cultures, Prédiction Agricole et Détection des maladies des plantes.

# **Fonctionnalités Principales**

# **1. Recommandation de Cultures**

recommandation.jpeg

Le Crop Recommendation Dataset utilisé est un ensemble de données qui contient des informations essentielles sur les sols et les conditions climatiques pour recommander des cultures agricoles spécifiques.

**Caractéristiques** utilisés :

 **•Teneur en Azote (N)**: Quantité d'azote dans le sol (kg/ha).

 **• Teneur en Phosphore (P)**: Quantité de phosphore dans le sol (kg/ha).

 **• Teneur en Potassium (K)**: Quantité de potassium dans le sol (kg/ha).

 **• Température (temperature)** : Température ambiante moyenne (°C)

 **• Humidité (humidity)**: Taux d'humidité de l'air ambiant (%).

 **• PH du Sol**: PH du sol, qui indique son niveau d’acidité ou d’alcalinité.

 **• Pluviométrie (rainfall)** : La quantité de précipitations (mm).

 **• Culture Recommandée (label)**: Culture la mieux adaptée à ces conditions,
      en fonction des besoins agronomiques et climatiques.


**Modèles de Machine Learning** testés:

 **• Random Forest**.

 **• Support Vector Machine**

 **• K-Nearest Neighbors**

 **• Gaussian Naïve Bayes**

 **• XGBoost**


Meilleur Modèle : **Gaussian Naïve Bayes** (Precision 99 %)

# **2. Prédiction Agricole**Prediction.jpg



La base de données que nous allons utiliser est constituée des rendements des dix cultures les plus consommées dans le monde et provient originellement du site web de l'Organisation des Nations Unies pour l'alimentation et l'agriculture (FAO).

**Caractéristiques** Utilisées :

 **• Area** : Pays (région où se trouve le champ)

 **• Item** : Le type de culture

 **• Year** : Année

 **• hg/ha_yield** : Rendement(hg/ha)

 **• average_rain_fall** : Precipitations moyennes (mm/an)

 **• pesticides_tonnes** : Pesticides utilisées (T)

 **• avg_temp** : Température moyenne (°C)

Modèles de Machine Learning testés:

**• Logistic Regression**

**• Random Forest**


**• Support Vector Machine**

**• Decision Tree**


Meilleur Modèle : **Random Forest**

Erreur Quadratique Moyenne : 18.24

Score R² : 0.973

# **Détection des Maladies des Plantes**

futuristic-businessman-farms-vegetables-crops-using-modern-ai-technology-mobile-phones-temperature-humidity-sensors-water-263535870.webp


AgriVision s’appuie sur les données et les technologies de PlantVillage, une plateforme de référence dans le domaine de la détection des maladies des plantes.Grâce à l'IA cette technologie permet de detecter en temps réel les maladies des plantes offrant ainsi aux agriculteurs des solutions rapides pour préserver la santé des plantes et éviter la propagation des infections.

**• Modèle** : **Yolov5**

• Entraîné sur divers jeux de données de maladies agricoles

• Classification précise des maladies à partir d'images

**Precision** : 83 %

**Recall** : 80.13 %



# **Deploiement**

AgriVision sera déployé via **Streamlit**, une solution légère et interactive permettant d’offrir une interface utilisateur intuitive pour la détection des maladies, la prédiction des rendements et les recommandations agricoles. Grâce à Streamlit, les utilisateurs pourront facilement télécharger des images de leurs cultures, obtenir des diagnostics en temps réel et accéder aux traitements recommandés, le tout depuis une plateforme accessible sur le web.
